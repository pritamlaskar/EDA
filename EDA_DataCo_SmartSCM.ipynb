```python
data.shape

# As data has 151 columns, we will make a function to identify columns with multi-collinearity (corr >= 0.75) and drop them
# Where columns size is significantly smaller, we can use a heatmap to do the same
```

```python
# Correlation matrix to identify columns with correlation >= 0.75
# Note: As a first-time coder, I have taken help from online sources (mainly StackOverflow and ChatGPT) for this matrix
# I used this matrix to learn and especially to directly remove columns with corr() > 0.75 as no. of columns is very high (151)
# As this matrix is difficult to make, we can also use VIF directly and eliminate columns with VIF > 5


def high_corr_columns(data):
    corr_matrix = data.corr().abs()
    high_corr_vars = []
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if corr_matrix.iloc[i, j] >= 0.75:
                colname = corr_matrix.columns[i]
                if colname not in high_corr_vars:
                    high_corr_vars.append(colname)
                    
    data = data.drop(high_corr_vars, axis = 1)
    return data


# Explanations:

# corr_matrix = data.corr().abs() : Calculates the correlation matrix and takes absolute values of correlation coefficients using .abs().
# high_corr_vars = [] : Empty list to store names of columns with corr() >= 0.75

# Rest of the code:

# These lines loop through the elements in the upper triangle of the correlation matrix (excluding the diagonal), and 
# (contd) checks if the absolute value of correlation coefficient is greater than or equal to 0.75. 
# (contd) If it is, the column associated with that coefficient is added to the high_corr_vars list, but only if it hasn't already been added.

# data = data.drop(high_corr_vars, axis = 1) : This line drops the high_corr_vars that had columns with corr() > = 0.75.
```

```python
# Applying the above correlation maxtrix function in our dataframe

data = high_corr_columns(data)
```

```python
data.shape
```

```python
# Now import VIF from statsmodel

from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data['feature'] = data.columns
```

```python
vif_data['VIF'] = [variance_inflation_factor(data.values, i) for i in range(len(data.columns))]
vif_data
```

```python
# Make a list of columns from vif_data where VIF > 5

cols_VIF_morethan5 = vif_data[vif_data["VIF"] > 5]["feature"].tolist()
cols_VIF_morethan5
```

```python
# Drop these columns

data = data.drop(cols_VIF_morethan5, axis = 1)
```
