```python
# Import seaborn, matplotlib, and numpy.

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Question 1: How does where you live affect your salary?

a = df.groupby(['employee_residence'])['salary_in_usd'].mean().sort_values(ascending = False).round(2)
print(a)
print()
print('Yes, where we live affect our salary. Salary is highest for employees living in in MY, PR, US, and lowest in UA, KE, IR.')
```

```python
# Question 2: How has the demand of the jobs been throughout the years?

a = df.groupby(['work_year'])['job_title'].count().sort_values(ascending = True)
print(a)
print()
print('Demand of jobs is growing as we can see that jobs have increased from 72 in 2020 to 318 in 2022.')
```

```python
# Question 3: How common is to work remote?

a = df.groupby(['work_year'])['remote_ratio'].mean()
print(a)
print()
print('The remote ratio is inreasing over the years and it is becoming more common.')
```

```python
# Question 4: What is the highest salary of EN and SE? Print salary along with job titiles and categories.

x = df[(df['experience_level']=="EN") | (df['experience_level']=="SE")]
result = x.groupby(["experience_level",'job_title'])['salary_in_usd'].max().sort_values(ascending=False)
print(result)
print('Highest salary in SE is Data Scientist of 412,000 USD and highest salary in EN is Machine Learning Engineer of 250,000 USD.')
```

```python
# Question 5: What company size hire the most?

a = df.groupby(['company_size'])['work_year'].count().sort_values(ascending = False)
print(a)
print()
print('M size zompanies hired the most.')
```

```python
# Question 6: How has average salary changed throughout the years?

a = df.groupby(['work_year'])['salary_in_usd'].mean().sort_values(ascending = True)
print(a)
print('Salary has increased minutely between 2020 and 2021 and by greatly between 2021 and 2022.')
```

```python
# Question 7: What are most popular roles in Data Science ?

sorted_job_title = df.groupby(['job_title'])['job_title'].count().sort_values(ascending = False).index

a = plt.figure(figsize = (15,10))
sns.countplot(data = df, y = 'job_title', order = sorted_job_title)
print()
print('Data Scientist is the most popular role in Data Science.')
```

```python
# Question 8: Which country hire the most people in Data Science?

a = df.groupby(['company_location'])['company_location'].count().sort_values(ascending=False).head(5)
print(a)
print()
print('US has hired the most people in Data Science')
```

```python
# Question 9: What is the distribution of Salaries?

a = sns.distplot(df['salary_in_usd'])
print()
print('Distribution of Salary is right skewed and most salaries lie between 0 USD and 200,000 USD.')
```

```python
# Question 10: How much can you expect depending on your years of experience?

a = df.groupby(['experience_level'])['salary_in_usd'].median().sort_values(ascending = False)
print(a)
print()
print('With four years of experience, one can expect 76,940 USD given his/her job is under MI category.')
```

```python
# Question 11: Which year do people prefer to stay at home the most?

a = df.groupby(['remote_ratio'])['work_year'].mean()
print(a)
print()
print('In the year 2021 most people have prefered to stay at home as the remote_rato is 100.')
```

```python
# Question 12: Which country has the highest pay?

a = df.groupby(['company_location'])['salary_in_usd'].mean().sort_values(ascending = False).round(2).head(5)
print(a)
print()
print('Russia has highest mean pay.')
```

```python
# Question 13: Which job title has the highest pay?

a = df.groupby(['job_title'])['salary_in_usd'].mean().sort_values(ascending = False).round(2).head(5)
print(a)
print()
print('Data Analytics Lead has the highest pay.')
```

```python
# Question 14: Is freelancing worth it?

a = df.groupby(['employee_residence', 'experience_level', 'job_title', 'remote_ratio'])['salary_in_usd'].agg(['mean']).sort_values(by = 'mean', ascending = False).head(5)
print(a)
print()
print('Freelancing is only worth if: a) Employee lives in US b) Employee experience is either EX or MI c) Employee job title is either Principal Data Engineer, Financial Data Analyst, or Principal Data Scientist as only these jobs have 100 remote ratio.')
```
